```python
import yfinance as yf

# Fetch stock data for Apple
ticker = "AAPL"
stock_data = yf.Ticker(ticker)

# Get historical market data
apple_data = stock_data.history(period="1y")  # Options: "1d", "5d", "1mo", "1y", etc.

# Display the first few rows
apple_data.head()
```
